* Área del círculo:
A = π · r²
Mide la superficie plana que ocupa el círculo. Se obtiene elevando el radio al cuadrado y multiplicándolo por el número pi (π ≈ 3.1416).
* 📏 Perímetro del círculo (Circunferencia):
P = 2 · π · r
Representa la longitud total del borde del círculo. Se calcula multiplicando el diámetro (2r) por pi.
* 🌍 Volumen de la esfera:
V = (4/3) · π · r³
Indica el espacio tridimensional que ocupa una esfera. Se obtiene elevando el radio al cubo, multiplicándolo por pi y por la fracción 4/3.
El programa está diseñado con funciones independientes para cada cálculo, un menú interactivo y estructuras de control básicas (while, if, elif, else). No utiliza programación orientada a objetos ni validación de datos, manteniéndolo sencillo y directo, ideal para un primer acercamiento a Python.


```python
# PROGRAMACION DE CALCULOS GEOMETRICOS
# elaborado por Axel Guzman
# formulas area del circulo, perimetro del circulo, volumen de la esfera


import math

def areacirculo():
    """Calcula y muestra el área de un círculo."""
    print("\n📐 Calculando el Área del Círculo")
    radio = float(input("Introduce el radio del círculo: "))
    area = math.pi * (radio ** 2)
    print(f"✅ El área del círculo con radio {radio} es: {area:.2f} unidades²\n")

def perimetrocirculo():
    """Calcula y muestra el perímetro de un círculo."""
    print("\n📏 Calculando el Perímetro del Círculo")
    radio = float(input("Introduce el radio del círculo: "))
    perimetro = 2 * math.pi * radio
    print(f"✅ El perímetro del círculo con radio {radio} es: {perimetro:.2f} unidades\n")

def volumenesfera():
    """Calcula y muestra el volumen de una esfera."""
    print("\n🌍 Calculando el Volumen de la Esfera")
    radio = float(input("Introduce el radio de la esfera: "))
    volumen = (4/3) * math.pi * (radio ** 3)
    print(f"✅ El volumen de la esfera con radio {radio} es: {volumen:.2f} unidades³\n")

def menu():
    """Muestra el menú interactivo y dirige el flujo del programa."""
    print("✨ ¡Bienvenido al Calculador Geométrico! ✨")
    while True:
        print("\n📋 MENÚ PRINCIPAL")
        print("1️⃣  Calcular área del círculo")
        print("2️⃣  Calcular perímetro del círculo")
        print("3️⃣  Calcular volumen de la esfera")
        print("4️⃣  Salir")

        opcion = input("\nSelecciona una opción (1-4): ")

        if opcion == "1":
            areacirculo()
        elif opcion == "2":
            perimetrocirculo()
        elif opcion == "3":
            volumenesfera()
        elif opcion == "4":
            print("\n👋 ¡Gracias por usar el programa! Hasta pronto. 🚀")
            break
        else:
            print("⚠️ Opción no válida. Por favor, elige un número del 1 al 4.")
            print("programa realizado por Axel Guzman 0070")

# Punto de entrada del programa
if __name__ == "__main__":
    menu()
```
